```python
import json
import paho.mqtt.client as mqtt
import requests

# ThingSpeakì˜ API Key
APIKEY = '---'

#  MQTT ì—°ê²° ì •ë³´
BROKER = 'damoa.io'
PORT = 1883
TOPIC = 'ewha/2391017'

# -----------------------------
# MQTT ì½œë°± í•¨ìˆ˜
# -----------------------------
def on_connect(client, userdata, flags, rc, properties=None):
    print("âœ… MQTT ì„œë²„ ì—°ê²° ì„±ê³µ")
    client.subscribe(TOPIC)
    print(f"ğŸ“¡ í† í”½ êµ¬ë… ì¤‘: {TOPIC}")

def on_message(client, userdata, msg):
    print(f"\nğŸ“¥ ìˆ˜ì‹ : {msg.topic}")
    data = msg.payload.decode('utf-8')
    print("RAW:", data)
    doit(data)

# -----------------------------
# ThingSpeak ì†¡ì‹  í•¨ìˆ˜
# -----------------------------
def doit(data):
    try:
        j = json.loads(data)
        # ë‹¤ì–‘í•œ í‚¤ ì´ë¦„ ëŒ€ì‘ (ì˜¨ë„, temp, t ë“±)
        t = j.get("ì˜¨ë„") or j.get("temp") or j.get("t")
        h = j.get("ìŠµë„") or j.get("hum") or j.get("h")

        if t is None or h is None:
            print("âš ï¸ ì˜¨ë„/ìŠµë„ ê°’ì´ ì—†ìŒ. ê±´ë„ˆëœ€.")
            return

        # ThingSpeak ì „ì†¡ URL
        url = f"https://api.thingspeak.com/update?api_key={APIKEY}&field1={t}&field2={h}"
        print("â¡ï¸ ì „ì†¡ URL:", url)

        r = requests.get(url)
        print("ì‘ë‹µ:", r.text)
    except Exception as e:
        print("âš ï¸ ë°ì´í„° ì²˜ë¦¬ ì˜¤ë¥˜:", e)

# -----------------------------
# MQTT í´ë¼ì´ì–¸íŠ¸ ì„¤ì •
# -----------------------------
client = mqtt.Client(mqtt.CallbackAPIVersion.VERSION2)
client.on_connect = on_connect
client.on_message = on_message

print(f"ğŸ”Œ {BROKER}:{PORT} ì—°ê²° ì‹œë„ ì¤‘...")

client.connect(BROKER, PORT)

client.loop_forever()
```
